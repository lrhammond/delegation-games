```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range of a and b
a_range = np.linspace(0, 1, 100)
b_range = np.linspace(0, 1, 100)

# Calculate the multiplication of a and b
c = a_range * b_range

# Plot the result
plt.plot(a_range, c)
plt.xlabel('a')
plt.ylabel('a * b')
plt.title('Multiplication of a and b')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Define values for V and H in the range [0, 1]
V = np.linspace(0, 1, 100)
H = np.linspace(0, 1, 100)

# Calculate C = V * H
H, V = np.meshgrid(H, V)
C = V * H

# Create a new figure and set its size
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Create a 3D scatter plot of C, V, and H
ax.scatter(V, H, C, s=40, c=C, cmap='plasma')

# Set the title and axis labels
ax.set_title('3D Scatter Plot of Coordination')
ax.set_xlabel('Vertical Coordination')
ax.set_ylabel('Horizontal Coordination')
ax.set_zlabel('Total Coordination')

# Show the plot
plt.show()
```
